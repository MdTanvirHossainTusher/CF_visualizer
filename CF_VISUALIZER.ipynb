```python
# Fill NaN values in Difficulty column with overall mean. this code is for if there is only one record of a paticular category.
# in our case, category 'O' has only one record with NAN difficulty. So, if we replace this NAN value with mean() of that category, the mean
# of the NAN value will also be NAN. To solve this, we find the overall mean and replace NAN of all category with this overall mean.
# Then again we override the Mean difficulty with category mean.

overall_mean_difficulty = df['Difficulty'].mean()
df['Difficulty'].fillna(overall_mean_difficulty, inplace=True)

# Calculate the mean difficulty for each category
category_mean_difficulty = df.groupby('Problem Category')['Difficulty'].mean()
category_mean_difficulty
```

```python
# Fill NaN values in Difficulty column with overall mean. this code is for if there is only one record of a paticular category.
# in our case, category 'O' has only one record with NAN difficulty. So, if we replace this NAN value with mean() of that category, the mean
# of the NAN value will also be NAN. To solve this, we find the overall mean and replace NAN of all category with this overall mean.
# Then again we override the Mean difficulty with category mean.

overall_mean_difficulty = df['Difficulty'].mean()
df['Difficulty'].fillna(overall_mean_difficulty, inplace=True)


## ami chacchi "Difficulty" column er NAN value gula replace krbo corresponding prb categoryr mean diye
## say, "Difficulty" column e F2 prb er difficulty NAN ase. Ami chai 'F2' er baki prb gular difficultyr mean oi NAN value te bosaite
## same idea goes for 'Total solve' column's NAN value

category_mean_difficulty = df.groupby('Problem Category')['Difficulty'].mean()


# Function to fill NaN values in Difficulty column with category mean
def fill_nan_with_category_mean(row):
    category = row['Problem Category']
    difficulty = row['Difficulty']
    if pd.isna(difficulty):
        return category_mean_difficulty[category]
    else:
        return difficulty

# Fill NaN values in Difficulty column with category mean
df['Difficulty'] = df.apply(fill_nan_with_category_mean, axis=1)

# Display the modified DataFrame
df
```

```python
# Filter and print the rows with NaN values
nan_rows = df[df.isnull().any(axis=1)]
nan_rows
```

```python
category_mean_solve = df.groupby('Problem Category')['Total solve'].mean()

# Function to fill NaN values in Difficulty column with category mean
def fill_nan_for_total_solve_with_category_mean(row):
    category = row['Problem Category']
    total_solves = row['Total solve']
    if pd.isna(total_solves):
        return category_mean_solve[category]
    else:
        return total_solves

# Fill NaN values in Difficulty column with category mean
df['Total solve'] = df.apply(fill_nan_for_total_solve_with_category_mean, axis=1)

# Display the modified DataFrame
df.tail(50)
```
